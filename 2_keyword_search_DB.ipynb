```python
from databricks.vector_search.client import VectorSearchClient

INDEX_NAME = "tdis_dev_data_catalog.tdir.rag_chunks_vs"

vsc = VectorSearchClient(disable_notice=True)
index = vsc.get_index(index_name=INDEX_NAME)
```

```python
def full_text_search(query_text: str, k: int = 50, columns=None):
    # Full-text keyword search (Beta): no embeddings used
    if columns is None:
        columns = ["chunk_id", "text", "source_file", "chunk_index_in_file"]
    res = index.similarity_search(
        query_text=query_text,
        columns=columns,
        num_results=k,
        query_type="FULL_TEXT"
    )
    return res
```

```python
def hybrid_search(query_text: str, k: int = 50, columns=None):
    # Hybrid keyword+vector search: BM25 + embedding
    if columns is None:
        columns = ["chunk_id", "text", "source_file", "chunk_index_in_file"]
    res = index.similarity_search(
        query_text=query_text,
        columns=columns,
        num_results=k,
        query_type="hybrid"
    )
    return res
```

```python
q = "Harris County flood damage estimate methodology"
```

```python
ft = full_text_search(q, k=20)
```

```python
hy = hybrid_search(q, k=20)
```
