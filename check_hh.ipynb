```python
from pyspark.sql import functions as F

base = "abfss://tdis-data-bronze@tdisproddatalakehouse.dfs.core.windows.net/RAG_files/"

# chunk_id for the join; source_file kept as just the file name
hh = (
    spark.read.json(base + "optimized_HH_chunks/*/chunks.jsonl")
    .select(
        "chunk_id",
        F.regexp_replace("source_file", r"^.*/", "").alias("source_file"),
    )
)
hh.createOrReplaceTempView("hh")

print("HH chunks in blob:", hh.count())
display(hh.select("source_file").distinct())
```

```python
%sql
SELECT
  (SELECT count(*) FROM hh)                                                    AS hh_total,
  (SELECT count(*) FROM hh JOIN tdis_dev_data_catalog.tdir.optimized_chunks_text               t USING (chunk_id)) AS in_text,
  (SELECT count(*) FROM hh JOIN tdis_dev_data_catalog.tdir.optimized_embeddings_dmretriever33m e USING (chunk_id)) AS in_emb,
  (SELECT count(*) FROM hh JOIN tdis_dev_data_catalog.tdir.optimized_rag_chunks               r USING (chunk_id)) AS in_rag_chunks
```
